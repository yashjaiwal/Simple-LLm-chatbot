```python
## example one simple LLm call 

from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage ,SystemMessage


```

```python
model = init_chat_model("groq:llama-3.1-8b-instant")
model
```

```python
### messagess

message = [
    SystemMessage("You are a helpful ai assiestent"),
    HumanMessage("hey what are the top 2 benefits of using Lanchain?")
]
```

```python
## invoke the model
response = model.invoke(message)
response
```

```python
print(response.content)
```

```python
#streaming example
for chunk in model.stream(message):
    print(chunk.content,end = "",flush= True)
```

```python
## dynamic promt templet

from langchain_core.prompts import ChatPromptTemplate

translation_template = ChatPromptTemplate.from_messages([
    ("system","You are a professional Translator. Translate the following text {text} from {source_language} to {target_language}.Maintain the tone and style"),
    ("user","{text}")
])

#using the templet
prompt = translation_template.invoke({
    "source_language": "English",
    "target_language":"Hindi",
    "text":"Langchain makes building AI application incredibly easy"

})
```

```python
prompt
```

```python
translated_response = model.invoke(prompt)
print(translated_response.content)
```

```python
# building first chai 
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough,RunnableLambda

def creat_story_chain():
    story_prompt = ChatPromptTemplate.from_messages([
        ("system","You are a creative storyteller. write a short and engaging story base on a given theme, character and setting"),
        ("user","theme:{theme}\n Main Character:{character}\n Settings:{setting}")
    ])

    #for story annalysis

    analysis_prompt = ChatPromptTemplate.from_messages([
        ("system","You are a literary critic.Analyze the following story and provide insights."),
        ("user","{story}")
    ])

    story_chain = (
    story_prompt | model | StrOutputParser()
    )
    def analyse_story(story_text):
        return {"story":story_text}

    analysis_chain = (
        story_chain
        | RunnableLambda(analyse_story)
        | analysis_prompt
        | model
        | StrOutputParser()
    )

    return analysis_chain
```

```python
chain = creat_story_chain()
chain
```

```python
result = chain.invoke({
    "theme": "Artificial Intelligence",
    "character": "a curious robot",
    "setting": "a futuristic city"
})

print(result)
```
